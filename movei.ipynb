```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. The small sample data from your original photo
data = {
    'Review': [
        'Boring and not interesting', 
        'Lacks good storyline and feels dull',
        'Comedy scenes are not funny',
        'Excellent movie without a dull moment', 
        'Great acting and feel good storyline'
    ],
    'Sentiment': ['Negative', 'Negative', 'Negative', 'Positive', 'Positive']
}

df = pd.DataFrame(data)

# 2. Setup the "Explainable" AI model
vectorizer = TfidfVectorizer(stop_words='english')
classifier = MultinomialNB()

# Train it
X = vectorizer.fit_transform(df['Review'])
classifier.fit(X, df['Sentiment'])

print("Success! Model is trained and ready.")
```

```python
def analyze_my_review(text):
    # Predict Sentiment
    vec = vectorizer.transform([text])
    prediction = classifier.predict(vec)[0]
    
    # Extract "Reasons" (Words the model knows)
    words = text.lower().split()
    features = vectorizer.get_feature_names_out()
    reasons = [w for w in words if w in features]
    
    print(f"Movie Review: {text}")
    print(f"Result: {prediction}")
    print(f"Reasoning Keywords: {reasons}")

# TEST IT HERE:
analyze_my_review("The acting was great but it was boring")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 1. Load the big dataset you just uploaded
df = pd.read_csv('IMDB-Dataset.csv')

# 2. Split into Training (80%) and Testing (20%)
X_train, X_test, y_train, y_test = train_test_split(
    df['review'], df['sentiment'], test_size=0.2, random_state=42
)

# 3. Vectorize (Limit to 5000 words so it runs fast)
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# 4. Train the Model
classifier = MultinomialNB()
classifier.fit(X_train_vec, y_train)

# 5. Check Accuracy (This is your key Viva stat!)
y_pred = classifier.predict(X_test_vec)
print(f"Model Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```
